# cleaning v_2

## Menghapus koma pada setiap kolom ID

```python
# daftar kolom ID yang perlu dibersihkan
id_cols = [
    'id', 'user_id', 'developer_id', 'submitter_id',
    'exam_registration_id', 'examinees_id', 'exam_module_id',
    'quiz_id', 'tutorial_id', 'journey_id'
]

def clean_ids(df):
    for col in df.columns:
        if col in id_cols and df[col].dtype == 'object':
            df[col] = df[col].astype(str).str.replace(',', '').str.strip()
    return df
```

```python
df_users_clean  = clean_ids(df_users_clean)
df_tracking_clean = clean_ids(df_tracking_clean)
df_submission_clean = clean_ids(df_submission_clean)
df_complete_clean = clean_ids(df_complete_clean)
df_regist_clean = clean_ids(df_regist_clean)
df_exam_clean = clean_ids(df_exam_clean)
```

## Drop kolom yang tidak dibutuhkan

```python
# Dictionary berisi dataframe dan kolom yang ingin di-drop
drop_columns = {
    "df_users_clean": [
        'remember_token', 'image_path', 'custom_city', 'verified_identity_document',
        'verified_certificate_name', 'user_role',
        'user_verification_status', 'created_at', 'updated_at', 'deleted_at',
        'city', 'city_id', 'unsubscribe_link', 'tz', 'verified_at', 'ama',
        'phone_verification_status', 'phone_verified_with'
    ],

    "df_tracking_clean": [
        'developer_journey_status_hash'
    ],

    "df_submission_clean": [
        'app_link', 'app_comment', 'admin_comment', 'note', 'current_reviewer',
        'pass_auto_checker'
    ],

    "df_complete_clean": [
        'updated_at'
    ],

    "df_regist_clean": [
        'updated_at'
    ],

    "df_exam_clean": [
        'look_report_at'
    ]
}

# Loop untuk menghapus kolom yang ada
for df_name, cols in drop_columns.items():
    df = globals()[df_name]
    df.drop(columns=[c for c in cols if c in df.columns], inplace=True)
    print(f"âœ” {df_name} cleaned â€” {len(cols)} columns checked.")

```

## Memmastikan Konsistensi TIpe Data

```python
# Ensure ID-like columns become string
id_columns = ['id', 'user_id', 'developer_id', 'submitter_id', 'quiz_id', 'tutorial_id',
              'journey_id', 'exam_module_id', 'examinees_id', 'exam_registration_id']

for df in [df_users_clean, df_tracking_clean, df_submission_clean, df_complete_clean, df_regist_clean, df_exam_clean]:
    for col in df.columns:
        if col in id_columns:
            df[col] = df[col].astype(str).str.replace(",", "").str.strip()
```

## Mengonversi Numeric Column

```python
numeric_cols = ['status', 'rating', 'study_duration', 'total_questions', 'score', 'is_passed']

for df in [df_submission_clean, df_complete_clean, df_exam_clean, df_tracking_clean]:
    for col in numeric_cols:
        if col in df.columns:
            df[col] = pd.to_numeric(df[col], errors='coerce')
```

## Memastikan Kolom Datetime Sudah Sesuai

```python
def convert_datetime(df):
    datetime_keywords = ["created_at", "updated_at", "deleted_at", "finished_at", "deadline_at", "started_at", "ended_at", "exam_finished_at"]

    for col in df.columns:
        if any(key == col.lower() for key in datetime_keywords):
            df[col] = pd.to_datetime(df[col], errors="coerce")

    return df
```

```python
df_users_clean = convert_datetime(df_users_clean)
df_tracking_clean = convert_datetime(df_tracking_clean)
df_submission_clean = convert_datetime(df_submission_clean)
df_complete_clean = convert_datetime(df_complete_clean)
df_regist_clean = convert_datetime(df_regist_clean)
df_exam_clean = convert_datetime(df_exam_clean)
```

# Menyimpan Dataset yang Sudah bersih

```python
df_users_clean.to_csv("clean_users.csv", index=False)
df_tracking_clean.to_csv("clean_tracking.csv", index=False)
df_submission_clean.to_csv("clean_submissions.csv", index=False)
df_complete_clean.to_csv("clean_completions.csv", index=False)
df_regist_clean.to_csv("clean_exam_registrations.csv", index=False)
df_exam_clean.to_csv("clean_exam_results.csv", index=False)

print("Dataset Hasil Cleaning Sudah DIsimpan")
```
