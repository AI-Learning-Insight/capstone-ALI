```python
import pandas as pd

# Load hasil feature engineering
features = pd.read_csv('features_ready.csv')

# Load dataset utama (buat ambil identitas user)
df = pd.read_csv('merged_cleaned.csv')

print("âœ… Data loaded")
print("features shape:", features.shape)
print("df shape:", df.shape)
```

```python
# Ambil kolom identitas dari df (tanpa duplikat)
user_info = df[['developer_id', 'display_name', 'email', 'user_role', 'city_id']].drop_duplicates()

# Gabungkan dengan data fitur
dashboard_data = pd.merge(
    features,
    user_info,
    on='developer_id',
    how='left'
)

# Cek hasil
dashboard_data.head()
```

```python
# Simpan hasil final untuk tim web
dashboard_data.to_csv('dashboard_data.csv', index=False)

print("âœ… dashboard_data.csv berhasil disimpan di folder data/processed/")
print("ðŸ“Š Ukuran data:", dashboard_data.shape)
print("ðŸ§  Kolom:", dashboard_data.columns.tolist())
```
