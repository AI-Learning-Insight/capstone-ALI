```python
import pandas as pd

# === 1Ô∏è‚É£ Load hasil feature engineering yang sudah dibersihkan ===
features = pd.read_csv('features_ready_cleaned.csv')

# === 2Ô∏è‚É£ Load dataset utama (buat ambil identitas user, seperti nama & kota) ===
df = pd.read_csv('merged_cleaned.csv')

print("‚úÖ Data loaded")
print("features shape:", features.shape)
print("df shape:", df.shape)

# (Opsional) Cek kolom kunci untuk memastikan bisa di-merge
print("\nKolom fitur:", features.columns.tolist())
print("Kolom df utama:", df.columns.tolist())
```

```python
# === 3Ô∏è‚É£ Pilih kolom identitas unik per developer ===
user_info = df[['developer_id', 'display_name', 'email', 'user_role', 'city_id']].drop_duplicates(subset=['developer_id'])

# === 4Ô∏è‚É£ Merge dengan hasil feature engineering ===
dashboard = pd.merge(features, user_info, on='developer_id', how='left')

# === 5Ô∏è‚É£ Cek hasil ===
print("‚úÖ Dashboard data created:")
print("Shape:", dashboard.shape)
print("\nKolom:", dashboard.columns.tolist())
print("\nContoh data:")
print(dashboard.head())
```

```python
dashboard.to_csv("dashboard_data.csv", index=False)
print("‚úÖ dashboard_data.csv berhasil disimpan! Siap dikirim ke tim web üöÄ")
```
